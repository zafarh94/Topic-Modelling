```python
import seaborn as sns
x1 = data.loc[data.Outcome=='Successful', 'Model_2']
x2 = data.loc[data.Outcome=='Unsuccessful', 'Model_2']

# Plot
kwargs = dict(hist_kws={'alpha':.6}, kde_kws={'linewidth':2})
# kwargs = dict(alpha=0.5, bins=100, density=True, stacked=True)

plt.figure(figsize=(10,7), dpi= 70)
sns.distplot(x1, color="Orange", label="Successful", **kwargs)
sns.distplot(x2, color="gray", label="Unsuccessful", **kwargs)
plt.xlim(0,1)

plt.legend();
```

```python
x1 = data.loc[data.Outcome=='Successful', 'Model_2']
x2 = data.loc[data.Outcome=='Unsuccessful', 'Model_2']
kwargs = dict(alpha=0.5, bins=100, density=True, stacked=True)
plt.hist(x1, **kwargs, color='orange', label='Successful')
plt.hist(x2, **kwargs, color='gray', label='Unsuccessful')
plt.gca().set(title='Frequency Histogram of Diamond Depths', ylabel='Probability')
plt.xlim(0,1)
plt.ylim(0,100)
plt.legend();
```
